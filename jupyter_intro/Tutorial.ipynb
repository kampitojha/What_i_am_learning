## ðŸ“Š Charts & Graphs (The Real Power)
Jupyter sabse zyada use hota hai **Data Analysis** ke liye kyunki ye charts wahin dikha deta hai.
Chalo ek graph banate hain!

```python
import matplotlib.pyplot as plt
import numpy as np

# Kuch fake data banate hain
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Graph plot karte hain
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Sine Wave', color='purple', linewidth=2)
plt.title('My First Jupyter Graph ðŸ“‰')
plt.legend()
plt.grid(True)
plt.show()
```
